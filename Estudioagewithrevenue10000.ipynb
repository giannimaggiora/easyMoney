```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
# from matplotlib import pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
import plotly.graph_objects as go
from plotly import offline as pyoff
```

```python
df=pd.read_csv('definitivo_10000clientes.csv')
```

```python
df['age_range'] = pd.qcut(df['age'], q=5, labels=range(5))
```

```python
df['age_range'].head()
```

```python
df.pivot_table(index='age_range', values=['monetary_value'], aggfunc=[len,np.mean, np.sum])
```

```python
df.head(40)
```

```python
plot_data = [
    go.Bar(
        x=df['age_range'],
        y=df['monetary_value'],
    )
]

plot_layout = go.Layout(
        xaxis={"type": "category"},
        title='Revenue por age range10000'
    )
colors = ['lightslategray']
plt.figure(figsize=(8,8))
fig = go.Figure(data=plot_data, layout=plot_layout)
fig.write_html('fig511.html')
pyoff.iplot(fig)
```
