```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
df=pd.read_csv('C:/Users/evive/Downloads/Nico y Juan/easyMoney/df_banc_rev.csv')
```

```python
df_region_code_unidades_vendidas = df.groupby('region_code')['unidades_vendidas_total'].sum().reset_index()
```

```python
df_region_code_unidades_vendidas.sort_values(by = ["unidades_vendidas_total"], ascending = False, inplace = True)
```
