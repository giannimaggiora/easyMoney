```python
for i in range(0, df.shape[0]):
    if(df.iloc[i]['segment_01 - TOP'] == 1):
        dfnew = dfnew.append({'segment': 1}, ignore_index=True)
        continue
    if(df.iloc[i]['segment_02 - PARTICULARES'] == 1):
        dfnew = dfnew.append({'segment': 2}, ignore_index=True)
        continue
    if(df.iloc[i]['segment_03 - UNIVERSITARIO'] == 1):
        dfnew = dfnew.append({'segment': 3}, ignore_index=True)
        continue
    print('break')
    break
```

```python
df_2 = pd.concat([df_d, dfnew], axis=1, join='inner')
```

```python
df_2.columns
```

```python
df_2['entry_channel'].value_counts()
```

```python
df_2.drop(['segment_01 - TOP','segment_02 - PARTICULARES', 'segment_03 - UNIVERSITARIO'], axis=1, inplace=True)
```

```python
df_2_entry_channel_unidades_vendidas = df_2.groupby('entry_channel')['unidades_vendidas_total'].sum().reset_index()
```

```python
plot_data = [
    go.Bar(
        x=df_2_entry_channel_unidades_vendidas['entry_channel'],
        y=df_2_entry_channel_unidades_vendidas['unidades_vendidas_total'],
    )
]

plot_layout = go.Layout(
        xaxis={"type": "category"},
        title='Unidades vendidas por entry channel'
    )

fig = go.Figure(data=plot_data, layout=plot_layout)
pyoff.iplot(fig)
```

```python
df_2_entry_channel_revenue = df_2.groupby('entry_channel')['revenue_total'].sum().reset_index()
```

```python
df_2_entry_channel_revenue.sort_values(by = ["revenue_total"], ascending = False, inplace = True)
```
